```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file_path = r"C:\Users\Admin\Downloads\DataScience_salaries_2025.csv"  # Fixed path
df = pd.read_csv(file_path)

print(df.head())
df.info(), df.head()
```

```python
df.dropna(inplace=True)
df['work_year'] = df['work_year'].astype(str)
```

```python
# Salary distribution visualization
plt.figure(figsize=(10, 5))
sns.histplot(df['salary_in_usd'], bins=50, kde=True, color='blue')
plt.title("Distribution of Data Science Salaries (USD)")
plt.xlabel("Salary in USD")
plt.ylabel("Frequency")
plt.show()
```

```python
# Experience level vs salary
plt.figure(figsize=(8, 5))
sns.boxplot(x='experience_level', y='salary_in_usd', data=df, palette='Set2')
plt.title("Salary Distribution by Experience Level")
plt.xlabel("Experience Level")
plt.ylabel("Salary in USD")
plt.show()
```

```python
# AI-related job salaries vs other data science roles
df_ai = df[df['job_title'].str.contains("AI|Machine Learning|Deep Learning", case=False)]
df_non_ai = df[~df['job_title'].str.contains("AI|Machine Learning|Deep Learning", case=False)]

plt.figure(figsize=(10, 5))
sns.boxplot(x=['AI-related Jobs']*len(df_ai) + ['Other Data Science Jobs']*len(df_non_ai),
            y=pd.concat([df_ai['salary_in_usd'], df_non_ai['salary_in_usd']]), palette='coolwarm')
plt.title("AI-related Jobs vs Other Data Science Salaries")
plt.ylabel("Salary in USD")
plt.show()
```

```python
# Remote work impact on salary
plt.figure(figsize=(8, 5))
sns.boxplot(x='remote_ratio', y='salary_in_usd', data=df, palette='muted')
plt.title("Remote Work Ratio vs Salary")
plt.xlabel("Remote Ratio (%)")
plt.ylabel("Salary in USD")
plt.show()
```

```python
# Salary by company size
plt.figure(figsize=(8, 5))
sns.boxplot(x='company_size', y='salary_in_usd', data=df, palette='pastel')
plt.title("Company Size vs Salary")
plt.xlabel("Company Size")
plt.ylabel("Salary in USD")
plt.show()
```

```python
#salary market insights
# Convert necessary columns to appropriate data types
df['work_year'] = df['work_year'].astype(str)

# Salary Distribution by Job Title
plt.figure(figsize=(12, 6))
top_jobs = df.groupby('job_title')['salary_in_usd'].median().sort_values(ascending=False)[:10]
sns.barplot(x=top_jobs.values, y=top_jobs.index, palette='coolwarm')
plt.title("Top 10 Highest-Paying Job Titles in Data Science")
plt.xlabel("Median Salary in USD")
plt.ylabel("Job Title")
plt.show()
```

```python
# Salary Trends Over the Years
plt.figure(figsize=(10, 5))
sns.lineplot(data=df, x='work_year', y='salary_in_usd', estimator='median', ci=None, marker="o")
plt.title("Median Salary Trends in Data Science (Over Years)")
plt.xlabel("Year")
plt.ylabel("Median Salary in USD")
plt.show()
```

```python
# Effect of Education on Salary
plt.figure(figsize=(10, 5))
sns.boxplot(x='experience_level', y='salary_in_usd', data=df, palette='Set2')
plt.title("Salary Distribution by Experience Level")
plt.xlabel("Experience Level")
plt.ylabel("Salary in USD")
plt.show()
```

```python
# Salary by Industry
df['industry'] = df['company_size'].map({'S': 'Small', 'M': 'Medium', 'L': 'Large'})
plt.figure(figsize=(10, 5))
sns.boxplot(x='industry', y='salary_in_usd', data=df, palette='pastel')
plt.title("Salary by Company Size")
plt.xlabel("Company Size")
plt.ylabel("Salary in USD")
plt.show()
```
